```python
!pip install xgboost
from os import environ, path
import pandas as pd
import xgboost

from sklearn.model_selection import train_test_split, KFold, GridSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
cleaned_data = pd.read_csv(path.join('C:/Users/Harish Vasanth/Desktop/Machine-learning-projects/DeliveryPredictionModel/data/', 'cleaned_data.csv'), index_col=0).reset_index(drop=True)
cleaned_data.drop(columns=['date','created_at','actual_delivery_time','date.1'])
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']

cleaned_data = cleaned_data.select_dtypes(include=numerics)
cleaned_data
cleaned_data.dropna()
```

### Base model 1

```python
y = cleaned_data.reindex(columns=['actual_time_taken'])

x = cleaned_data.drop(columns=['actual_time_taken'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, shuffle=True, random_state=0)
```

```python
model = xgboost.XGBRegressor()
history = model.fit(
    x_train, y_train,
    eval_set=[(x_test, y_test)],
    eval_metric=['rmse', 'mae'], 
    early_stopping_rounds=10)
```
